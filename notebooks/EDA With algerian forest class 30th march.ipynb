## Data cleaning

```python
dataset[dataset.isnull().any(axis = 1)]
```

The dataset is converted into two sets based on region from 122th index 
we can make a new column based on the region 

1. Bejaia Region Dataset

2. Sidi-Bel Abbes Region dataset

Add new column with region

```python
dataset.loc[:122 , "Region"] = 0
dataset.loc[122: , "Region"]=1
df = dataset
```

```python
df.info()
```

```python
df.head(5)
```

```python
df['Region'] = df['Region'].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Changes the required columns as integer data type

```python
df.columns
```

```python
df[['day' , 'month' , 'year' , 'Temperature', 'RH' , 'Ws']] = df[['day' , 'month' , 'year' , 'Temperature', 'RH' , 'Ws']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

## Changing the other columns to float datatype

```python
objects = [feature for feature in df.columns if df[feature].dtype == 'O']
```

```python
objects
```

```python
for i in objects:
    if i != 'Classes' :
        df[i] = df[i].astype(float)
```

# Exploratory Data Analysis

```python
# Drop day, month, year
df_copy = df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head(5)
```

```python
# Categories in classes
df_copy['Classes'].value_counts()
```

```python
# Encoding of the categories in classes 
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
# plot density plot for all features

plt.style.use('seaborn')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
# Percentage for pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# Plotting pie chart
classlabels = ['FIRE' , 'NOT FIRE']
plt.figure(figsize=(15,10))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%',colors='bg')
plt.title('Pie chart of classes')
plt.show()
```

```python
# Monthly fire analysis

dftemp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(15,6))
sns.set_style('whitegrid')
sns.countplot(x = 'month' , hue='Classes' , data=dftemp)
plt.ylabel('Number of fires' , weight = 'bold')
plt.xlabel('Months' , weight = 'bold')
plt.title('Fire Analysis of Sidi-Bel Regions' , weight = 'bold')

```

```python
# Monthly Fire analysis

dftemp = df.loc[df['Region']==0]
plt.subplots(figsize=(15,6))
sns.set_style('whitegrid')
sns.countplot(x = 'month' , hue = 'Classes' , data=dftemp)
plt.xlabel('Months' , weight='bold')
plt.ylabel('Number of Fires' , weight = 'bold')
plt.title('Fire Analysis of Bejaia Region' , weight = 'bold')
```

1. IT is observed that july and August had the most number of forest fires for both the regions. and from the above plot of months we can understand few things,

2. Most of the fires happend in August and and very high Fires happened in only 3 months - June , July and August in Bejaia Region.

3. Less Fire was on September in Bejaia Region.

4. Most of the fires happend in August and and very high Fires happened in only 3 months - July, August and September in Sidi-Bel Region.

5. Less fire was on June in Sidi-Bel Region.

6. Sidi-Bel Region has more fires when compare to Bejaia Region.
